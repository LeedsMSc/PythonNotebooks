```python
# Import modules
from pathlib import Path
import easygui


class FindFiles():
    """
    Open a  file dialogbox to allow a "parent" folder to be selected.  The folder is searched for a requested file type 
    and the identified files/paths added to a list.  If required the search can be performed recursively. 
    
    -----------------------------------------------------------------------
    # Attributes:
    
        recursive_search = boolean (recursive or non-recursive)
            
        file_type = the file extension to search
               
    -----------------------------------------------------------------------
        
    """

    def __init__(self, recursive_search = True, file_type = "xls[mx]",):
        # Recursive and file_type
        self.recursive_search = recursive_search
        self.file_type = file_type
        
        # Target Folder and file paths
        self.target_folder = self.select_a_folder()
        self.files_list = list(self.file_paths())
        
        # Print out file paths to notepad
        self.print_files()
        

    def select_a_folder(self):
        """
        Select "parent" folder and get path as object.
        """
        target_path = easygui.diropenbox(title="Select a folder", default=Path.home().joinpath(*["OneDrive"]))
        try:
            self.target_folder = Path(target_path)
            return(self.target_folder)
        except:
            self.target_folder = ""
            return(self.target_folder)
    

    def file_paths(self):
        """
        Get files from target folder
        """
        if self.target_folder == "": print("No folder selected"); return None
        self.files_list = self.target_folder.rglob("*." + self.file_type) if self.recursive_search else self.target_folder.glob("*." + self.file_type)
        return(self.files_list)
        
        
    def print_files(self):
        """
        Print out files list.
        """
        with open('files found.txt', 'w') as f:
            for file in self.files_list:
                f.write(f'Path(r"{file}"), \\' + '\n') 
```
